## NLTK

- Python上著名的⾃然语⾔处理库
- ⾃带语料库，词性分类库
- ⾃带分类，分词，等等功能
- 强⼤的社区⽀持
- 还有 N 多的简单版 wrapper

### NLTK ⾃带语料库

```
# 自带的 corpus 语料库集合，导入 brown （大学的）语料库
In [2]: from nltk.corpus import brown

# 里面包含的很多分类
In [3]: brown.categories()
Out[3]:
['adventure',
 'belles_lettres',
 'editorial',
 'fiction',
 'government',
 'hobbies',
 'humor',
 'learned',
 'lore',
 'mystery',
 'news',
 'religion',
 'reviews',
 'romance',
 'science_fiction']

In [4]: len(brown.sents())
Out[4]: 57340

In [5]: len(brown.words())
Out[5]: 1161192


```
